```python
import pandas as pd
# Cargar el archivo
df = pd.read_csv('/content/country_vaccinations.csv')
#Estruct DataFrame
print(df.info())
print('------------------')
#COnversion columna
df['date'] = pd.to_datetime(df['date'])
#agrupar
vacunas_comp = df.groupby('vaccines')['total_vaccinations'].sum()
print(vacunas_comp)
print('------------------')
#total vacunas
total_vacunas= df['total_vaccinations'].sum()
print(total_vacunas)
print('------------------')
# Promedio por pais
promedio_pais = df.groupby('country')['total_vaccinations'].mean()
print(promedio_pais)
print('------------------')
#Vacunas 29 enero
vacunas_29E = df[df['date'] == '2021-01-29']['daily_vaccinations'].sum()
print(vacunas_29E)
print('------------------')
#Nueva columna con difs
df['diferencias'] = df['daily_vaccinations'] - df['daily_vaccinations_raw']
conDiferencias = df.copy()
print(conDiferencias.head())
print('------------------')
#Periodo de tiempo
periodo_t = df['date'].max() - df['date'].min()
print(periodo_t)
print('------------------')
# Crear una nueva columna con la cantidad de vacunas utilizadas cada día
df['canVac'] = df['vaccines'].str.split(',').apply(len)
conCantidad = df.copy()
print(conCantidad.head())
print('------------------')
# Filtrar los registros antes del 20 de diciembre de 2020
antes20 = df[df['date'] < '2020-12-20']
print(antes20.head())
print('------------------')
#FIltrar por pfizer
pfizer = df[df['vaccines'].str.contains('Pfizer')]
print(pfizer.head())
print('------------------')
with pd.ExcelWriter('resultadosReto.xlsx') as writer:
    conDiferencias.to_excel(writer, sheet_name='conDiferencias')
    conCantidad.to_excel(writer, sheet_name='conCantidad')
    antes20.to_excel(writer, sheet_name='antes20')
    pfizer.to_excel(writer, sheet_name='pfizer')
```
